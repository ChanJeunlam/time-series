```matlab

kappa=zeros(1,2);lambda=zeros(1,2);theta=zeros(1,2);
for i=1:2
    if i==1
        cvi=vfilt(rcv(:,4),24,'mirror');
    elseif i==2
        cvi=rcv(:,4)-vfilt(rcv(:,4),24,'mirror');
    end
    u=real(cvi);u=u-mean(u);
    v=imag(cvi);v=v-mean(v);
    [a2,b2,theta(i)]=specdiag(mean(u.*u),mean(v.*v),mean(u.*v));
    [kappa(i),lambda(i)]=ab2kl(sqrt(a2),sqrt(b2));
end

set(gcf,'paperposition',[1 1 10 6])   %set suitable figure size
for i=1:2
    subplot(1,2,i)
    %this is the same code block we used for the plot just above, 
    %but for the ith depth instead of the 4th depth
    if i==1
       cvi=vfilt(rcv(:,4),24,'mirror');
    elseif i==2
        cvi=rcv(:,4)-vfilt(rcv(:,4),24,'mirror');
    end
   [mat,xmid,ymid]=...
        twodhist(real(cvi),imag(cvi),[-50:50],[-50:50]); %the last two arguments set the x- and y- bin sizes
    mat(mat==0)=nan; %swap zero values in the histogram for NaNs, so they don't appear with a color
    jpcolor(xmid,ymid,mat);hold on
    colormap lansey, axis equal, axis(49*[-1 1 -1 1])
    vlines(0,'k:'),hlines(0,'k:')
    
        h=colorbar('South');
        h.Label.String='Observations per bin';
        
    plot([0+0*1i mean(cvi)],'color','w','linewidth',3) %ploting a white edge around the line for visibility
    plot([0+0*1i mean(cvi)],'color','k','linewidth',2)
    plot(mean(cvi),'wo','markerfacecolor','k')
    xtick([-40:10:40]),ytick([-40:10:40])

    %now we add a plot of the variance ellipse
    h=ellipseplot(kappa(i),lambda(i),theta(i),mean(cvi));%jlab code to plot ellipses
    set(h,'color','w'),set(h,'linewidth',4)
    h=ellipseplot(kappa(i),lambda(i),theta(i),mean(cvi));%jlab code to plot ellipses
    set(h,'color',[1 1 1]*0.6),set(h,'linewidth',2)

    text(-40,40,['Depth = ' int2str(depths(4)) ' m'])
    caxis([0 80])
end

#from matplotlib.patches import Ellipse  #Our patch for plotting the variance ellipse



#ellipse plot with built ins

#xo=np.mean(cvr.real[:,3])
#yo=np.mean(cvr.imag[:,3])
#ellipse = Ellipse((xo,yo), width=2*a, height=2*b, angle=np.rad2deg(theta),fill=False,linewidth=4.5,color="w")
#ax.add_patch(ellipse)
#ellipse = Ellipse((xo,yo), width=2*a, height=2*b, angle=np.rad2deg(theta),fill=False,linewidth=3,color=(0.6,0.6,0.6))
#ax.add_patch(ellipse)


#convolution using signal processing toolbox

#fcvr =  sg.convolve(cvr[:,3], window, mode="same")  #filtered version at deepest depth

#ft =  sg.convolve(t-np.mean(t.data), window, mode="same") + np.mean(t.data)#filtered version at deepest depth
#In the above I'm adding and subtracting the mean to minimize edge effcts




print(np.size(lons))
     
dlon=lons_gc[1]-lons_gc[0] 
dlat=(lats_gc[1]-lats_gc[0])/2+(lats_gc[-1]-lats_gc[-2])/2
lons_gc=lons_gc-dlon/2  
lats_gc=lats_gc-dlat/2



lons_gc=topo_ds["lon"][lona_index:lonb_index+1] #add one extra longitude 
lats_gc=topo_ds["lat"][lata_index:latb_index+1] #add one extra latitude
dlon=lons_gc[1]-lons_gc[0]
dlat=(lats_gc[1]-lats_gc[0])/2+(lats_gc[-1]-lats_gc[-2])/2
lons_gc=lons_gc-dlon/2  
lats_gc=lats_gc-dlat/2



#phi=np.arange(0,2*np.pi+.1,.1)
#z=np.exp(1j*phi)
#z=np.exp(1j*theta)*(a*np.cos(phi)-1j*b*np.sin(phi))

def plot_twodhist(xbins,ybins,x,y,tickstep=False,plotdim=(1,1),figsize=plt.rcParams.get('figure.figsize'),axlines=(0,0),logN=False,cmap = sns.color_palette("Spectral_r", as_cmap=True),meandot=True,meanline=False,axisequal=False):
    """
    Compute and plot two dimensional histogram.
    
    Args: 
        xbins: Array of bin edges for x-bins
        ybins: Array of bin edges for y-bins
        x: Array of x-values to be binned
        y: Array of y-values to be binned; same size as x
        
    Optional Args:
        tickstep: X- and y-axis tick step, a length 2 tuple, defaults to auto   
        plotdim: Number of rows and columns, a length 2 tuple, defaults to (1,1)
        figsize: Figure size, a length 2 tuple, defaults to system default
        axlines: Axis origin locations for horizontal and vertical lines,
            a length 2 tuple, defaults to (0,0), lines omitted if False
        logN: Plot logarithmic rather than linear values in bins  
            logN=10 uses log10, logN=2 uses log2
        cmap: Colormap, defaults to Spectral_r
        meandot: Plots a dot at the mean value, defaults to true
        meanline: Plots a line from the origin to the mean value, defaults to false
        axisequal: Sets plot aspect ratio to equal, defaults to false

    Returns:
        ax: Axis handle
        im: Image handle; an integer when the figure contains a single subplot, 
            and a list of integers otherwise.       
    """
    
    def plot_twodhist_one(axn,xn,yn):
        #plot just one twodhist
        z = stats.binned_statistic_2d(xn,yn, None, bins=[xbins, ybins], statistic="count")
        z.statistic[z.statistic==0]=np.nan  #swap zero values for NaNs, so they don't appear with a color
        
        if logN==2:
            imn=axn.pcolormesh(xbins, ybins, np.log2(np.transpose(z.statistic)), cmap=cmap, shading="flat")
            cb=fig.colorbar(imn, ax=axn)
            cb.set_label('Log2 Number of Observations')
        elif logN==10:
            imn=axn.pcolormesh(xbins, ybins, np.log10(np.transpose(z.statistic)), cmap=cmap, shading="flat")
            cb=fig.colorbar(imn, ax=axn)
            cb.set_label('Log10 Number of Observations')
        else:
            imn=axn.pcolormesh(xbins, ybins, np.transpose(z.statistic), cmap=cmap, shading="flat")
            cb=fig.colorbar(imn, ax=axn)
            cb.set_label('Number of Observations')
    
        if axisequal:
            axn.set_aspect("equal")
        
        if not(not(axlines)):
            axn.axhline(axlines[0], linestyle=":", color="k")
            axn.axvline(axlines[1], linestyle=":", color="k")

        if meanline:
            #plt.arrow(0,0,np.mean(x),np.mean(y),width=0.8,length_includes_head=False,facecolor="k",edgecolor="w")
            axn.plot([0,np.mean(x)],[0,np.mean(y)],color="w",linewidth=4.5)
            axn.plot([0,np.mean(x)],[0,np.mean(y)],color="k",linewidth=3)
            
        if meandot:
            axn.plot(np.mean(x),np.mean(y), "wo", markerfacecolor="k", markersize=10)

        plt.xlim([min(xbins), max(xbins)]),plt.ylim([min(ybins), max(ybins)])

        if not(tickstep==False):
            plt.xticks(np.arange(min(xbins), max(xbins), tickstep[0]))  # set x-label locations
            plt.yticks(np.arange(min(ybins), max(ybins), tickstep[1]))  # set x-label locations
    
        return imn
    
    fig, ax = plt.subplots(plotdim[0], plotdim[1], figsize=figsize)  #sharex=True,sharey=True
    N=plotdim[0]*plotdim[1]
    
    if N==1:
        im=plot_twodhist_one(ax,x,y)
    else:
        im=[]  #initialize a list of image handles
        for n in np.arange(0,N):        
            im.append(plot_twodhist_one(ax.reshape(N)[n],x[:,n],y[:,n]))
       
    return ax,im

bins = np.arange(-50, 50.01, 1)
ax=plot_twodhist(bins,bins,cvr.real[:,3],cvr.imag[:,3],tickstep=(10,10),meanline=True,axisequal=True)
plt.xlabel('Downstream Velocity (cm/s)'),plt.ylabel('Cross-Stream Velocity (cm/s)')
plt.title('Two-Dimensional Histogram of Velocity');
```
